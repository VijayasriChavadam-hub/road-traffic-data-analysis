```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv("traffic_data.csv")

# Show data
print(df.head())

# Find busiest location
busiest = df.groupby("location")["vehicle_count"].sum()
print("Busiest Location:\n", busiest)

# Peak traffic time
peak_time = df.sort_values(by="vehicle_count", ascending=False)
print("Peak Traffic:\n", peak_time.head())

# Plot
df.groupby("time")["vehicle_count"].sum().plot(kind='bar')
plt.title("Traffic by Time")
plt.xlabel("Time")
plt.ylabel("Vehicle Count")
plt.show()
```
